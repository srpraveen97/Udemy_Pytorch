```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = torch.linspace(1.0,50.0,50).reshape(-1,1)
```

```python
e = torch.randn_like(X,dtype=torch.float)
```

```python
Y = 2*X + 30 + 10*e
```

```python
plt.scatter(X.numpy(),Y.numpy());
```

```python
model = nn.Linear(1,1)
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.001)
```

```python
epochs = 10000
losses = []

for i in range(epochs):
    i = i+1
    Y_pred = model.forward(X)
    loss = criterion(Y_pred,Y)
    losses.append(loss)
    print(f'epoch: {i:2}  loss: {loss.item():10.8f}  weight: {model.weight.item():10.8f}  \
bias: {model.bias.item():10.8f}')
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()    
```

```python
plt.plot(range(epochs), losses)
plt.ylabel('Loss')
plt.xlabel('epoch');
```

```python
w1,b1 = model.weight.item(), model.bias.item()
x1 = np.linspace(0.0,50.0,50)
y1 = x1*w1 + b1
```

```python
plt.scatter(X.numpy(), Y.numpy())
plt.plot(x1,y1,'r')
plt.title('Current Model')
plt.ylabel('y')
plt.xlabel('x');
```
